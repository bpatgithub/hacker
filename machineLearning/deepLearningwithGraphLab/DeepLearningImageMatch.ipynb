#### we will use K nearest neighbour algorithm.

```python
import graphlab
```

```python
# Limit number of worker processes. This preserves system memory, which prevents hosted notebooks from crashing.
graphlab.set_runtime_config('GRAPHLAB_DEFAULT_NUM_PYLAMBDA_WORKERS', 4)
```

```python
# load the data for images.

image_train = graphlab.SFrame('image_train_data/')
```

```python
# view data.
graphlab.canvas.set_target('ipynb')
image_train.show()
```

```python
# trian the model using nearest-neighbour algorithm.

knn_model = graphlab.nearest_neighbors.create(image_train,
                                             features=['deep_features'],
                                             label='id')
```

```python
# now let's pick 4th image as its cat as we show in above show() command.

test1=image_train[3:4]
test1['image'].show()
```

```python
# now let's find out similar images.
knn_model.query(test1)
```

```python
# we don't see the images, so that's not easy to understand.  

res = knn_model.query(test1)
```

```python
# now let's retrive the image.
image_train.filter_by(res['reference_label'],'id').show()
```

```python
# good results, but not perfect.
# let's create a function to make it easier to find neighbours and find out the quality of results.

def show_neighbours(i):
    res = knn_model.query(image_train[i:i+1])
    fil_res = image_train.filter_by(res['reference_label'],'id')
    fil_res['image'].show()
    
```

```python
show_neighbours(13)
```

```python
# nice.  First image is the query and it is matching with all cars with almost similar white color.
# let's play bit more.
```

```python
show_neighbours(1537)
```
