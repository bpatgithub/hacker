```python
# there are total 2005 records, with 509 autos, 509 cats, 509 dogs and 478 birds.
```

```python
dog_data = image_train[image_train['label'] == 'dog']
```

```python
auto_data = image_train[image_train['label'] == 'automobile']
```

```python
cat_data = image_train[image_train['label'] == 'cat']
```

```python
bird_data = image_train[image_train['label'] == 'bird']
```

```python
# let's review auto_data.

auto_data.show()
```

```python
# now let's create deep learning model for each of the categories.

dog_model = graphlab.nearest_neighbors.create(dog_data,
                                             features=['deep_features'],
                                             label='id')
```

```python
auto_model = graphlab.nearest_neighbors.create(auto_data,
                                             features=['deep_features'],
                                             label='id')
```

```python
cat_model = graphlab.nearest_neighbors.create(cat_data,
                                             features=['deep_features'],
                                             label='id')
```

```python
bird_model = graphlab.nearest_neighbors.create(bird_data,
                                             features=['deep_features'],
                                             label='id')
```

```python
# let's review the dog data and find the nearest neighbours.  Here is the first image in the dog data.

dq = dog_data[0:1]
dq['image'].show()
```

```python
# now let's find out similar images using deep learning.  Remember, it doesn't have lable any more.  
# In other words, it has same label, so not useful.

dog_model.query(dq)
```

```python
# let's create a function to make it easier to find neighbours and find out the quality of results.

def show_neighbours(model, data, i):
    res = model.query(data[i:i+1])
    fil_res = data.filter_by(res['reference_label'],'id')
    fil_res['image'].show()
    
```

```python
show_neighbours(dog_model, dog_data,0)
```

```python
show_neighbours(dog_model, dog_data,135)
```

## Now a big challenge.  Measure the accuracy among these different models.

```python
# first let's load the train data.

image_test = graphlab.SFrame('image_test_data/')
```

```python
# we already created training data for each category.  now let's create test data for each category.

dog_test_data = image_test[image_test['label'] == 'dog']
```

```python
auto_test_data = image_test[image_test['label'] == 'automobile']
```

```python
cat_test_data = image_test[image_test['label'] == 'cat']
```

```python
bird_test_data = image_test[image_test['label'] == 'bird']
```

```python
# in above section, we have found nearest neighbour for one image at a time.
# in this section, we will call model to find all nearest neighbour for each image.
```

```python
# First use dog model to find out nearest 1 neighbour.
# We will find the 1 (one) neighbour (k=1).
dog_dog_nbrs = dog_model.query(dog_test_data,k=1)
```

```python
# Use Cat Model on dog data.
# We will find the 1 (one) neighbour (k=1).
dog_cat_nbrs = cat_model.query(dog_test_data,k=1)
```

```python
# Use Auto Model on dog data.
dog_auto_nbrs = auto_model.query(dog_test_data,k=1)
```

```python
# Use Bird Model on dog data.
dog_bird_nbrs = bird_model.query(dog_test_data,k=1)
```

```python
dog_cat_nbrs.show()
```

```python
# now let's find out the distance for dog from dog data, cat data, auto data and bird data.

dog_distance = graphlab.SFrame({'dog_dog': dog_dog_nbrs['distance'],
                                'dog_cat': dog_cat_nbrs['distance'],
                                'dog_auto': dog_auto_nbrs['distance'],
                                'dog_bird': dog_bird_nbrs['distance'],
                               })
```

```python
dog_distance.head()
```

```python
# as you can see, dog_dog has the lowest distance which make sense.  
# but not all rows.  The last row, dog_cat has lower distance than dog_dog.
# That means, dog image was closer to cat than dog!
# let's find out how many times dog_dog wins over other images.

def dog_wins(row):
    if (
        (row['dog_dog'] - row['dog_cat']) <= 0 and
        (row['dog_dog'] - row['dog_bird']) <= 0 and
        (row['dog_dog'] - row['dog_auto']) <= 0
    ):
        return 1
    else:
        return 0
```

```python
dog_distance.apply(dog_wins).sum()
```
